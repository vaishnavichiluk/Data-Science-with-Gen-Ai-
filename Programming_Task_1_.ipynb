# Python If-Else

```python
n=int(input())
if (n%2==1):
    print('Weird')
elif n%2==0 and 2<=n<=5:
    print('Not Weird')
elif n%2==0 and 6<=n<=20:
    print('Weird')
elif n%2==0 and n>20:
    print('Not Weird')
```

# Write a function

```python
def is_leap(year):
    leap = False

    # Write your logic here
    if year%4==0 and (year%100!=0 or year%400==000):
        leap=True
    else:
        leap=False

    return leap

```

```python
print(is_leap(2025))
```
