# Running Sum of 1d Array

```python
class Solution:
    def runningSum(self, nums):
        n=len(nums)
        temp=[0]*n
        temp[0]=nums[0]
        for i in range(1,n):
            temp[i]=temp[i-1]+nums[i]

        return temp
```

```python
o=Solution()
print(o.runningSum([1,2,3,4]))
```

# Shuffle the Array

```python
class Solution:
    def shuffle(self, nums, n):
        n1=len(nums)
        right=nums[:n]
        left=nums[n:]
        temp=[]
        ri=0
        li=0
        for i in range(n1):
            if i%2==0 :
                temp.append(right[ri])
                ri+=1
            else:
                temp.append(left[li])
                li+=1

        return temp


```

```python
c=Solution()
print(c.shuffle([2,5,1,3,4,7],3))
```

# Kids With the Greatest Number of Candies

```python
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
        n=len(candies)
        m=float('-inf')
        temp=[]
        for i in range(n):
            if candies[i]>m:
                m=candies[i]

        for i in range(n):
            ec=candies[i]+extraCandies
            if ec>=m:
                temp.append(True)
            else:
                temp.append(False)
        return temp
```

```python
d=Solution()
print(d.kidsWithCandies([2,3,5,1,3],3))
```
